### Horizontal Bar ###

```python
plt.figure(figsize=(9,6))
plt.barh(y=df['Brand'], width=df['Cars Listings'],color='midnightblue')
plt.xticks(rotation=0)
plt.xlabel('Number of Listing',fontsize=13)
plt.title('Cars by Brand',fontsize=17)
```

```python
plt.figure(figsize = (9, 6)) 
# To create a horizontal bar chart with matplotlib you can use the 'barh' function.
plt.barh(y = df["Brand"], 
        width = df["Cars Listings"].sort_values(), 
        color = "midnightblue") 
plt.xticks(fontsize = 13) 
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold") 
plt.ylabel("Number of Listings", fontsize = 13 ) 
plt.savefig("Used Cars Bar.png") 
plt.show() 
```

### Lolipop Plot

```python
# Create a dataframe
df = pd.DataFrame({'country':['United States','Russia','France','Germany','Israel',
                            'United Kingdom','Netherlands',
                            'China','Spain','Italy','South Korea',
                            'Turkey','Ukraine','Switzerland','Norway'], 
                   'spending':np.random.uniform(size=15) })
df
```

```python
# Reorder it following the values:
ordered_df = df.sort_values(by='spending')
my_range=range(1,len(df.index)+1)

plt.hlines(y=my_range, xmin=0, xmax=ordered_df['spending'], color='skyblue')
plt.plot(ordered_df['spending'], my_range, "o")
 
# Add titles and axis names
plt.yticks(my_range, ordered_df['country'])
plt.title("A vertical lolipop plot", loc='center', fontsize=17)
plt.xlabel('Spending in Millions')
plt.ylabel('Countries')

plt.show()
```

### Heat Map ###

```python
cr_df = sns.load_dataset('car_crashes')
cr_df.columns
```

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)
# So we see that n previous accident is heavily correlated with accidents, while the
# insurance premium is not
crash_mx = cr_df.corr()

# Create the heatmap, add annotations and a color map
sns.heatmap(crash_mx, 
            annot=True, 
            cmap='Blues'
            #cmap='viridis'
           )
plt.show()
```
